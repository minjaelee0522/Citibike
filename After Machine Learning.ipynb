```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('../data/Most300_Stations.csv',index_col=0)
```

```python
columns = ['start station checkout counts', 'commute', 'Temperature','start hour','start station latitude',
           'start station longitude','start dayofweek', 'PRCP','start year','start month',
           'SNOW','usertype', 'gender', 'start day','Isweekday']
corr = df[columns].corr()
```

#  Correlation Between Features XGBoost

```python
sns.set(style="whitegrid")
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 15))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

# Rush Hours effect on Checkout Counts

```python
rushhour = df.groupby('commute').agg({'start station checkout counts':'mean'}).reset_index()
rushhour.columns = ['Rush Hours','Hourly Checkout Counts'] 
plt.figure(figsize = (10,4))
sns.set_style('darkgrid')
sns.barplot(x ='Rush Hours', y ='Hourly Checkout Counts', data = rushhour);
```

# Temperature effect on Checkout Counts

```python
temp = df.groupby(['start year','start month','start day'])[['start station checkout counts', 'Temperature']].mean().reset_index()
temp.columns=['Year','Month','Day','Hourly Checkout Counts','Temperature']
plt.figure(figsize = (10,4))
sns.set_style('darkgrid')
sns.scatterplot(x ='Temperature', y ='Hourly Checkout Counts', data = temp);
```

# Hour effect on Checkout Counts

```python
hours = df.groupby('start hour').agg({'start station checkout counts':'mean'}).reset_index()
hours.columns = ['Hours','Hourly Checkout Counts'] 
plt.figure(figsize = (15,8))
sns.set_style('darkgrid')
sns.barplot(x ='Hours', y ='Hourly Checkout Counts', data = hours);
```

# Day of Week effect on Checkout Counts

```python
day = df.groupby('start dayofweek').agg({'start station checkout counts':'mean'}).reset_index()
day.columns = ['Day Of Week','Hourly Checkout Counts'] 
plt.figure(figsize = (15,8))
sns.set_style('darkgrid')
sns.barplot(x ='Day Of Week', y ='Hourly Checkout Counts', data = day, palette='husl');
```

# Month effect on Checkout Counts

```python
month = df.groupby('start month').agg({'start station checkout counts':'mean'}).reset_index()
month.columns = ['Month','Hourly Checkout Counts']
plt.figure(figsize = (15,8))
sns.set_style('darkgrid')
sns.barplot(x ='Month', y ='Hourly Checkout Counts', data = month);
```
